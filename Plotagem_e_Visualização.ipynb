## Acrescentando legendas



1.   Plotagem simples com três linhas e uma legenda




```
from numpy.random import randn
```

```
#Plotagem simples com três linhas e uma legenda
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ticks = ax.set_xticks([0,250,500,750,1000]) #Definindo as "casas" do eixo X
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label="two")
ax.plot(randn(1000).cumsum(), 'k.', label="three")
ax.legend(loc='best') # Encontra o melhor lugar para alocar a legenda
```

## Anotações e desenhos em uma subplotagem



1.   Visualização de dados composta de três patches distintos
2.   Salvando plots em diversos formatos





```
fig = plt.figure()
ax = fig.add_subplot(1,1,1)


rect = plt.Rectangle((0.2,0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15],[0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
plt.savefig('figpath.png', dpi=400, bbox_inches='tight') #Salvando com melhor resolução
plt.savefig('figpath.svg')
plt.savefig('figpath.pdf')
```

#### importando dataset Gorjetas 

https://www.kaggle.com/leonardobolek/anlise-de-gorjetas

```
tips = pd.read_csv('tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts                          
```

```
tips
```

```
#Não há muitos grupos de 1 e de 6 pessoas

party_counts = party_counts.loc[:, 2:5]
```

```
#Normalize para que a soma seja 1 

party_pcts = party_counts.div(party_counts.sum(1), axis=0)
party_pcts
```

```
#Fração dos grupos por tamanho a cada dia
party_pcts.plot.bar()
```

# S E A B O R N 

```
import seaborn as sns 
```

```
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])
tips.head()
```

```
#Porcentagem de gorjetas por dia com barras de erro
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

```
# Porcentagem de gorjeta por dia e hora
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
sns.set(style="whitegrid")
```

## Plotagens de dispersão ou de pontos

```
macro = pd.read_csv('https://raw.githubusercontent.com/statsmodels/statsmodels/master/statsmodels/datasets/macrodata/macrodata.csv')
```

```
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data[-5:]
```

```
# Uma plotagem de regressão/dispersão do seaborn
sns.regplot('m1', 'unemp', data=trans_data)
plt.title('Titulo bonito aqui')
```

```
# Matriz de plotagens de pares dos dados macro de statsmodels
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```
